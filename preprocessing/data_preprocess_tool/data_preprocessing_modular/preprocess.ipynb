```python
import json
import re
import os
```

```python

def get_key_words_from_topic(f_adr):
    with open(f_adr,'r') as f:
        data = json.loads(f.read())
        titles = [ele['topic'] for ele in data]
        keyworddic = []
        for title in titles:
            keyworddic.append([word for word in title.split(' ') if not (re.match('\.\.\.',word) or re.match('[0-9]*/[0-9]*\n',word))])
        for ele,keywords in zip(data,keyworddic):
            ele['key_words'] = keywords
        for e in data:
            e['content'] = ' '.join([w for w in e['content'].split(' ') if w is not '\n'])
    return data

if __name__ == '__main__':
    
    for i in range(1,11):
        try:
            if i < 10:
                week = 'week0'
            else:
                week = 'week'
            if not os.path.exists('./train/'):
                os.makedirs('./train/')
            file_adr = './data_preprocessed/'+ week + str(i)+'.json'
            
            data = get_key_words_from_topic(file_adr)
            data = json.dumps(data, indent=1)
            with open(file_adr,'w') as f:
                f.write(data)
        except:
            print('week',i,'was not processed...')
```

```python
def add_train_data(data):
    ques = []
    sol  = []
    #code = []
    qa = []
    title = [' '.join(ele['key_words']) for ele in data]
    
    for t in title:
        ques.append("What is "+ t + " ?")
        ques.append("What does "+ t + " mean ?")
        ques.append("Tell me something about "+ t + " ?")
        ques.append(t)
        ques.append("Explain " + t +" ?")
        ques.append("Elaborate " + t +" ?")
        ques.append("Can you tell me about " + t + " ?")
        ques.append("What do you know about " + t + " ?")
        ques.append("What can you tell me about " + t + " ?")
        ques.append("I want to know about" + t )
        ques.append("Do you have information about " + t + " ?")
    for t in title:
        sol.append("How to solve "+ t + " ?")
        sol.append("How to resolve "+ t + " ?")
        sol.append("How to mitigate "+ t + " ?")
        sol.append("Solution for "+ t)
        sol.append("Provide me some Solution for "+ t)
        sol.append("mitigation for "+ t)
        sol.append("How to stop "+ t + " ?")
        sol.append("How to defend "+ t + " ?")
        sol.append("How to get secured against "+ t + " ?")
        sol.append("Solution, "+t)
    
    for que, ele in zip(ques,data):
        qa.append({'question': que,'answer': ele['content']})
    for so, ele in zip(sol,data):
        qa.append({'question': que,'answer': ele['content']})    

    return qa
```
